```python
import soundfile as sf
import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-large-xlsr-53-french")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-xlsr-53-french")
```

```python

tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
error_rate=[]
for file_name in test_french_wav:
  try:
    input_audio, _ = librosa.load(file_name,sr=16000)
    input_values = tokenizer(input_audio, return_tensors="pt").input_values
    logits = model(input_values).logits
    predicted_ids = torch.argmax(logits, dim=-1)
    predicted_transcription = tokenizer.batch_decode(predicted_ids)[0]
    f=file_name.replace('.wav','')
    f=f.replace("/content/drive/MyDrive/Projet Fin D'etude/Speech To text/french/","")
    #print(f)
    true_transcription=french.loc[french['file_name']==f]['tran'].values[0]
    error = wer(true_transcription, predicted_transcription)
    error_rate.append(error)
  except:
    continue
```

```python
error_rate
```

```python
error_rate=[]
for file_name in test_english_wav:
  input_audio, _ = librosa.load(file_name,sr=16000)
  input_values = tokenizer(input_audio, return_tensors="pt").input_values
  logits = model(input_values).logits
  predicted_ids = torch.argmax(logits, dim=-1)
  predicted_transcription = tokenizer.batch_decode(predicted_ids)[0]
  f = open(file_name.replace('wav','txt'), "r")
  true_transcription=f.read()
  error = wer(true_transcription, predicted_transcription)
  error_rate.append(error)
```

```python
from deepspeech import Model
import numpy as np
import os
import wave
import json
```

```python

model_file_path = 'deepspeech-0.8.2-models.pbmm'
lm_file_path = 'deepspeech-0.8.2-models.scorer'
beam_width = 100
lm_alpha = 0.93
lm_beta = 1.18

model = Model(model_file_path)
model.enableExternalScorer(lm_file_path)
```

```python
model.setScorerAlphaBeta(lm_alpha, lm_beta)
model.setBeamWidth(beam_width)
```

```python
error_rate1=[]
for file_name in test_english_wav:
  fin=wave.open(file_name, 'rb')
  audio = np.frombuffer(fin.readframes(fin.getnframes()), np.int16)
  sample_rate = fin.getframerate()
  predicted_transcription=model.stt(audio)
  f = open(file_name.replace('wav','txt'), "r")
  true_transcription=f.read()
  error = wer(true_transcription, predicted_transcription)
  error_rate1.append(error)
```

```python
sum(error_rate1)/len(error_rate1)
```

```python
def read_wav_file(filename):
    with wave.open(filename, 'rb') as w:
        rate = w.getframerate()
        frames = w.getnframes()
        buffer = w.readframes(frames)
    
    

    return buffer, rate
```

```python
wf = wave.open(file_name)
rate = wf.getframerate()
frames = wf.getnframes()
buffer = wf.readframes(frames)
print(len(buffer))
```

```python
read_wav_file(file_name)
```

```python
def transcribe_streaming(audio_file):
    buffer, rate = read_wav_file(audio_file)
    offset=0
    batch_size=8196
    text=''

    while offset < len(buffer):
      end_offset=offset+batch_size
      chunk=buffer[offset:end_offset]
      data16 = np.frombuffer(chunk, dtype=np.int16)

      stream.feedAudioContent(data16)
      text=stream.intermediateDecode()
      #clear_output(wait=True)
      print(text)
      offset=end_offset
    return True
```
